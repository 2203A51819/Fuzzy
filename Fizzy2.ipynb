```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import skfuzzy as fuzz
```

```python
x = np.arange(0, 40, 0.1)
```

```python
cold = fuzz.trapmf(x, [0, 0, 10, 15])
```

```python
warm = fuzz.trapmf(x, [10, 15, 25, 30])
```

```python
hot = fuzz.trapmf(x, [25, 30, 40, 40])
```

```python
plt.figure()
```

```python
plt.plot(x, cold, label='Cold')
```

```python
plt.plot(x, warm, label='Warm')
```

```python
plt.plot(x, hot, label='Hot')
```

```python
plt.xlabel('Temperature')
```

```python
plt.ylabel('Membership')
```

```python
plt.title('Fuzzy Membership Functions')
```

```python
plt.legend()
```

```python
plt.grid()
```

```python
plt.show()
```
