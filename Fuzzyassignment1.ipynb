```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import skfuzzy as fuzz
```

```python
speed = np.arange(0, 101, 1)
```

```python
slow = fuzz.trapmf(speed, [0, 0, 20, 30])
```

```python
medium = fuzz.trapmf(speed, [20, 30, 60, 70])
```

```python
fast = fuzz.trapmf(speed, [60, 70, 100, 100])
```

```python
plt.figure()
```

```python
plt.plot(speed, slow, label='Slow')
```

```python
plt.plot(speed, medium, label='Medium')
```

```python
plt.plot(speed, fast, label='Fast')
```

```python
plt.xlabel('Speed (km/h)')
```

```python
plt.ylabel('Membership')
```

```python
plt.title('Fuzzy Membership Functions for Speed Classification')
```

```python
plt.legend()
```

```python
plt.grid()
```

```python
plt.show()
```
