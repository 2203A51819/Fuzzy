```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 40, 0.1)
```

```python
cold = np.maximum(0, np.minimum(1, np.minimum((15-x)/(15-10),1)))
```

```python
warm = np.maximum(0,np.minimum(1, np.minimum((x-10)/(20-10),(30-x)/(30-25))))
```

```python
hot= np.maximum(0,np.minimum(1, np.minimum((x-25)/(30-25),0)))
```

```python
plt.figure(figsize=(12,6))
```

```python
plt.plot(x, cold, label='cold', linewidth=2)
```

```python
plt.plot(x, warm, label='warm', linewidth=2)
```

```python
plt.plot(x, hot, label='hot', linewidth=2)
```

```python
plt.xlabel('Temperature (*c)')
```

```python
plt.ylabel('Membership Degree(values of )')
```

```python
plt.title('Fuzzy Sets for Temperature Levels')
```

```python
plt.legend()
```

```python
plt.grid(True)
```

```python
plt.show()
```
